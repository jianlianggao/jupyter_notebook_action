```python
# sample code for testing 02 same image output
import matplotlib.pyplot as plt
import math, os
import numpy as np
from pathlib import Path
#import ipynbname
stem = "test002"
# Make rendering deterministic
plt.style.use("default")
plt.rcParams.update({
    "font.family": "DejaVu Sans",
    "figure.dpi": 100,
    "savefig.dpi": 100,
    "lines.antialiased": True,
    "path.simplify": False,
})

# Data (adjust to match the baseline you expect)
x = [i * 200 for i in range(1000)]   # <-- use 200 if baseline uses 200
y = [math.sin(i * 0.01) for i in range(1000)]

# Plot
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.plot(x, y)
ax.set_title("Sine plot")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Save with stable bounding box
Path("results").mkdir(exist_ok=True)
arr = np.column_stack([x, y])

np.savetxt(
    f"results/{stem}_xy.txt",
    arr,
    fmt=["%d", "%.17g"],
    delimiter="\t",
    header="x\ty",
    comments=""
)
fig.savefig(f"results/{stem}.png")  
print(f"{stem} saved to results/{stem}.png")
```
