```python
# sample code for testing 02 same image output
import matplotlib.pyplot as plt
import math
from pathlib import Path

x = [i * 200 for i in range(1000)]  # Baseline uses 200
y = [math.sin(i * 0.01) for i in range(1000)]

fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.plot(x, y)
ax.set_title("Baseline Sine Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")




# Save the figure so CI can compare it
Path("results").mkdir(exist_ok=True)
fig.savefig("results/test001.png", dpi=100, bbox_inches="tight")
print("Baseline saved to results/test001.png")
```
