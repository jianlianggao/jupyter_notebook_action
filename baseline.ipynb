```python
# sample code for baseline txt and png
import matplotlib.pyplot as plt
import math
import numpy as np
from pathlib import Path


# Make rendering deterministic
plt.style.use("default")
plt.rcParams.update({
    "font.family": "DejaVu Sans",
    "figure.dpi": 100,
    "savefig.dpi": 100,
    "lines.antialiased": True,
    "path.simplify": False,
})

# Data (adjust to match the baseline you expect)
x = [i * 200 for i in range(1000)]   # <-- use 200 if baseline uses 200
y = [math.sin(i * 0.01) for i in range(1000)]

# Plot
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)
ax.plot(x, y)
ax.set_title("Sine plot")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Save with stable bounding box
Path("baseline").mkdir(exist_ok=True)
arr = np.column_stack([x, y])
np.savetxt(
    "baseline/test000_xy.txt",
    arr,
    fmt=["%d", "%.17g"],
    delimiter="\t",
    header="x\ty",
    comments=""
)
fig.savefig("baseline/test000.png")  
print("Baseline saved to baseline/test000.png")
```
